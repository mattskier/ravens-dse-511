### Goal:  Perform exploratory data analysis (EDA)

This dataset contains financial information from a bank in Czech Republic about their clients and their accounts. There are eight tables available in the dataset.

### Dataset description:
This dataset has 8 tables: https://sorry.vse.cz/~berka/challenge/pkdd1999/berka.htm

- account: each record describes static characteristics of an account <br>
[account_id, district_id, date, frequency] - 4500 entries

- client: each record describes characteristics of a client <br>
[client_id, birth_number, district_id] 

- disposition: each record relates together a client with an account <br>
[disp_id, client_id, account_id, type] 

- permanent order: each record describes characteristics of a payment order <br>
[order_id, account_id, bank_to, account_to, amount, K_symbol]

- transaction: each record describes one transaction on an account <br>
[trans_id, account_id, date, type, operation, amount, balance, k_symbol]

- loan: each record describes a loan granted for a given account <br>
[loan_id, account_id, date, amount, duration, payments, status]

- credit card: each record describes a credit card issued to an account <br>
[card_id, disp_id, type, issued]

- demographic data:  each record describes demographic characteristics of a district <br>
[A1, ..., A16]

### Data Preprocessing/Cleaning

- handle missing values 
- convert NoneType to empty strings
- convert date to Datetime objects
- translate Czech terminology to English for ease of understanding
- convert birth date to age

```python
def check_for_any_missing_values(data):
    """
    This function checks for any missing values in a dataframe
    and return a boolean (True/False). 
    Args:
        data: pandas dataframe 
    """
    return data.isnull().values.any()
```

```python
# Helpful functions for handling data

# convert NoneType to empty string
xstr = lambda s: s or ' '   

# roughly calculate the age from birth date
convert_age = lambda t: dt.date.today().year - t.year    

# translate Czech to English
trans_type = lambda s: 'credit' if s == 'PRIJEM' else 'withdrawal' 
disp_type = lambda s: 'owner' if s == "OWNER" else 'user'

def trans_operation(x):
    if x == 'VYBER KARTOU':
        return 'credit card withdrawal'
    elif x == 'VKLAD':
        return 'credit in cash'
    elif x == 'PREVOD Z UCTU':
        return 'from another bank'
    elif x == 'VYBER':
        return 'withdrawal in cash'
    elif x == 'PREVOD NA UCET':
        return 'to another bank'
    else:
        return ' '

def trans_ksymbol(x):
    if x == "POJISTNE":
        return 'insurance'
    elif x == "SLUZBY":
        return 'statement'
    elif x == "UROK":
        return 'interest credited'
    elif x == "SANKC. UROK":
        return 'sanction interest'
    elif x == "SIPO":
        return 'household'
    elif x == "DUCHOD":
        return 'pension'
    elif x == "UVER":
        return 'loan'
    elif x == "LEASING":
        return 'leasing'
    else:
        return " "
    
def account_freq(x):
    if x == "POPLATEK MESICNE":
        return 'monthly'
    elif x == "POPLATEK TYDNE":
        return 'weekly'
    elif x == "POPLATEK PO OBRATU":
        return 'after transaction'
    else:
        return " "
    
def loan_status(x):
    if x == 'A':
        return 'contract finished, no problem'
    elif x == 'B':
        return 'contract finished, loan not paid'
    elif x == 'C':
        return 'running contract, OK so far'
    elif x == 'D':
        return 'running contract, client in debt'
    else:
        return ' '
```

```python
# Preprocessing data
account['date'] = pd.to_datetime( account['date'], format='%Y-%m-%d')
account['frequency'] = account['frequency'].apply(xstr).apply(account_freq)
card['issued'] = pd.to_datetime( account['date'], format='%Y-%m-%d')
client['birth_date'] = pd.to_datetime( client['birth_date'], format='%Y-%m-%d')
# client['age'] = client['birth_date'].apply(xstr).apply(convert_age)
trans = pd.read_sql_query(query_trans, conn)
trans['date'] = pd.to_datetime( account['date'], format='%Y-%m-%d')
trans['operation'] = trans['operation'].apply(xstr).apply(trans_operation)
trans['k_symbol'] = trans['k_symbol'].apply(xstr).apply(trans_ksymbol)
trans['bank'] = trans['bank'].apply(xstr)
trans['type'] = trans['type'].apply(trans_type)
loan['date'] = pd.to_datetime( account['date'], format='%Y-%m-%d')
order['k_symbol'] = order['k_symbol'].apply(trans_ksymbol)
disp['type'] = disp['type'].apply(disp_type)
# loan['status'].apply(loan_status)
```

### Summary:

The dataset is obtained from a Czech Republic bank, dated back from 1993-1997. It has 4500 accounts and 5369 clients. 
- Household payment is the majority of permanent order.
- The gender ratio of the clients is close to 1:1.
- Most frequent transaction mode is withdrawal in cash.
- Majority of the clients (about 90%) with a bank loan are in group C (running contract, OK so far) and group A (contract finished, no problems). The rest may be problematic.
- Most credit card type belongs to classic (about 74%)
- There are 5 regions and 77 districts (bank branches). Prague has the highest average salary and number of entrepreneurs per 1000 inhabitants.

```python
trans['operation'].replace('', np.nan, inplace=True)
trans.dropna(subset=['operation'], inplace=True)
trans_sorted = trans.sort_values('date')
trans_sorted['year'] = pd.DatetimeIndex(trans_sorted['date']).year
trans_sorted['month'] = pd.DatetimeIndex(trans_sorted['date']).month
trans_sorted['day'] = pd.DatetimeIndex(trans_sorted['date']).day
```

```python
# Find trend in mode of transaction over the years

plt.rcParams.update(params)
sns.countplot(data=trans_sorted, x='operation', hue='year', palette='Set1')
plt.title("Frequency of different modes of transaction")
plt.xlabel("Mode of transaction")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
```

Figure 7: Cash withdrawal is the most popular mode of transaction every year from 1993 to 1997. The breakdown of transaction modes is consistent over the years as well. Credit card withdrawal may have only started to gain some popularity starting 1996.

```python
# Track cash inflow and outflow and how it affects the transaction balance 

trans_gb = trans.groupby(['account_id', 'type'], as_index=False).mean()
g1 = sns.jointplot(data=trans_gb, x='trans_amount', y='trans_balance', hue='type', height=7)
```

Figure 8: The x-axis represents the amount of the transaction and the y-axis represents the balance of an account after a transaction. In the top plot, the red datapoints represents cash outflow (withdrawal) and the blue represents cash inflow (credit). 

The transaction amounts for both credit and withdrawal tend to fall within the range below 10,000 CZK. Transactions with amount greater than 10,000 CZK tend to be source of income. Transaction with credit type shows an upward trend where the large the amount of credit, the higher the account balance. Meanwhile, most clients kept the amount of withdrawal sufficiently low where they don't break their bank account (a downward trend in the account balance is not observed).

```python
# Perform 'join' on multiple tables: loan, demographic data, account
# to find out what client features/attributes results in different loan statuses

query_combined = "SELECT A2, A3, A11, A12, A13, A14, A15, A16, \
                 account.account_id,\
                 loan.loan_amount, loan.duration, loan.payments, loan.status \
                 FROM (district \
                 INNER JOIN account on district.district_id = account.district_id) \
                 INNER JOIN loan on account.account_id = loan.account_id"

combined = pd.read_sql_query(query_combined, conn)
combined = combined.rename(columns={'A2':'district name', 'A3':'region', 'A11': 'average salary',
                        'A12': 'unemployment_rate_95', 'A13': 'unemployment_rate_96', 
                        'A14': 'no_of_entrepreneurs_per_1000',
                        'A15': 'no_of_commited_crimes_95', 'A16': 'no_of_commited_crimes-96',
                        'amount': 'loan amount', 'duration': 'loan duration (month)', 
                        'payments': 'loan payment', 'status': 'loan status'
                        })

combined['No of crimes'] = combined['no_of_commited_crimes_95'] + combined['no_of_commited_crimes_95']
combined['unemployment rate'] = (combined['unemployment_rate_95'] + combined['unemployment_rate_96'])/2
combined.head()
```

```python
sns.countplot(data=combined, x='region', hue='loan status')
```

Figure 9: Most clients from different regions belong to either loan status 'A' or 'C', which indicate a good loan record track. North Moravia seems to have a higher percentage of clients with loan status 'B' and 'D'. It might be interesting to see what other attributes that might be causing that. 

```python
g = sns.catplot(data=combined, y="region", x='unemployment rate', orient='h', hue='loan status',
                kind='bar', height=10)
g.set_xticklabels(rotation=45)
```

```python
g = sns.catplot(data=combined, y="region", x='No of crimes', orient='h', hue='loan status',
                kind='bar', height=10)
g.set_xticklabels(rotation=45)
```

Figure 10: In all regions, clients with loan status 'D' has the highest amount of loan compared to other loan statuses. Loan status 'D' are running contracts where clients are in debt. In general, Prague tends to have large amount of loan , which may have some correlation with the higher average salary and number of entrepreneurs per 1000 inhabitants in Prague. With more entrepreneurs in the region, it is not surprising to see a large amount of loan form that region. 

Clients with loan status 'A' (contract finished with no debts) tend to have a loan amount of about 10,000 CZK or less. Clients with loan status 'C' have a loan amount ranges from 15,000 - 20,000 CZK. It would be interesting to see the loan status of a client based on the average salary.

```python
# Trend of crime rate and average salary in different regions 
g = sns.jointplot(data=combined.groupby("region").mean(), x='average salary', y='No of crimes', s=100,
                 alpha=.7, edgecolor=".2", linewidth=.5, height=8)
g.ax_joint.set_yscale('log')
```

Figure 11: It is interesting to see an positive trend in the number of crimes with increasing average salary. As we explored before that regions with higher average salary doesn't necessarily indicate a lower unemployment rate. Meanwhile, we observed that the unemployment rate seems to largely correlate with the number of crimes in the region. 

This suggests that the population economic/wealth disparity in different regions might not be well portrayed in this dataset.
